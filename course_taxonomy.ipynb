```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy

%matplotlib inline
```

```python
# Display all columns in pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
enroll_df = pd.read_csv('enrollment_data_combined.csv')
```

```python
# Fill nulls of the 'Subj' column with 'Dept'
enroll_df['Department'] = enroll_df[' Subj'].fillna(enroll_df[' Dept'])
```

```python
course_dept_counts = enroll_df.groupby('Title')['Department'].nunique().reset_index()
course_dept_counts = course_dept_counts.sort_values(by='Department', ascending=False)
```

```python
# Filter enroll_df to only include courses that appear in 2 or fewer departments - this still makes up 79% of all course enrollment data
filtered_enroll_df = enroll_df[enroll_df['Title'].isin(course_dept_counts[course_dept_counts['Department'] <= 2]['Title'])]
filtered_enroll_df.shape[0] / enroll_df.shape[0]
```
